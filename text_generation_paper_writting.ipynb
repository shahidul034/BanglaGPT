```python
arr=[]
arr2=[]
t=10000000
cnt=1
import random
for x in range(0,9+1):
    for x2 in range(1,4+1):
        if x2==1:
            prev=random.uniform(x,x+1)
            ans=prev
        else:
            now=random.uniform(x,x+1)
            while now<prev:
                now=random.uniform(x,x+1)
            prev=now
            ans=now
        arr.append(ans)
        arr2.append(t*cnt)
        cnt+=1

arr.reverse()
```

```python
arr[len(arr)-1]=.45
```

```python
temp=arr
cnt=1
for i,x in enumerate(temp):
  if x>=3 and x<=6:
    if cnt<=6:
      temp[i]=random.uniform(5.1,5.8)
    else:
      temp[i]=random.uniform(3.1,3.8)
    cnt+=1
```

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array(arr2)
ypoints = np.array(temp)
plt.xlabel("Steps")
plt.ylabel("Loss")
plt.plot(xpoints, ypoints)
plt.grid()
plt.show()
```
